Data preprocessing

```python
#outliers
outliers_col=['lat','long','city_pop','merch_lat','merch_long','is_fraud']
for i in outliers_col:
  Q1=df[i].quantile(0.25)
  Q3=df[i].quantile(0.75)
  IQR=Q3-Q1
  df[(df[i]>Q1-1.5*IQR)&(df[i]<Q3+1.5*IQR)]
  sns.boxplot([outliers_col])
  plt.xlabel(i)
  plt.ylabel('distribution')
  plt.show()
```

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor
x = df[num_col]
x = x.drop('is_fraud', axis=1)
x = x.select_dtypes(include=np.number)
x = x.replace([np.inf, -np.inf], np.nan)
x = x.fillna(x.mean())



def VIF(df):
    vif_data = pd.DataFrame()
    vif_data['feature'] = df.columns
    vif_data['VIF'] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]
    return vif_data

while True:
    vif_data = VIF(x)
    max_value = vif_data['VIF'].max()
    if max_value <= 5:
        break
    else:
        col_to_drop = vif_data.loc[vif_data['VIF'].idxmax(), 'feature']
        print(f"Dropping '{col_to_drop}' with VIF: {max_value}")
        x = x.drop(col_to_drop, axis=1)
```

Train_test_split

```python
x=df_final.drop('is_fraud',axis=1)
y=df_final['is_fraud']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.70,random_state=45)
```

Model building-Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini')
model.fit(x_train,y_train)
```

Model evaluation

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
cn_mt=confusion_matrix(y_test,y_pred)
accuracy=accuracy_score(y_test,y_pred)
clss_report=classification_report(y_test,y_pred)
```

```python
print("Accuracy",accuracy)
print("Classification_Report",clss_report)
print("Confusion_matrix",cn_mt)
```
